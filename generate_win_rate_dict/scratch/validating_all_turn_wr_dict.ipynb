## Checked that there weren't duplicate cards or duplicate turn dicts for that card 1/4/2024

```python
import os 
import sys
import pandas as pd
import json
os.chdir('..\\..')
```

```python
# checking out card_wr_by_turn_dict/card_wr_by_turn_total.json
path_to_json = r'data\card_wr_by_turn_dict\card_wr_by_turn_total.json'
with open(path_to_json, 'r') as f:
    card_wr_by_turn_total = json.load(f)
```

```python
# check if every key is unique
all_keys = []
duplicate_keys = []
for key in card_wr_by_turn_total.keys():
    if key in all_keys:
        duplicate_keys.append(key)
    else:
        all_keys.append(key)

print(len(duplicate_keys))
print(len(all_keys) == len(card_wr_by_turn_total.keys()))
```

```python
for key in card_wr_by_turn_total.keys():
    unique_keys = []
    print('working on card:', key)
    for turn_key in card_wr_by_turn_total[key].keys():
        print('   working on turn:', turn_key)
        if turn_key in unique_keys:
            print(f'{key} has duplicate keys')
            break
        else:
            unique_keys.append(turn_key)

```
